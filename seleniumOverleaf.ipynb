```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome(executable_path=r"chromedriver.exe")
driver.maximize_window()


# Navigate to the application home page
driver.get("https://pt.overleaf.com/")

#clicando no botão de login
btt_login = driver.find_elements_by_xpath("/html/body/nav/div/div[2]/ul/li[6]/a")[0]
btt_login.click()

# get the search textbox
username = driver.find_element_by_name("email")
username.clear()

password = driver.find_element_by_name("password")
password.clear()

# enter search keyword and submit
username.send_keys("saulo.moreira@gec.inatel.br")
password.send_keys("1996.sjm")
username.submit()
password.submit()

#clicando em new project
wait = WebDriverWait(driver, 10)
wait.until(EC.visibility_of_element_located((By.XPATH, "/html/body/main/div[2]/div/div/div[1]/aside/div[1]/a")))

#view all
btt_novoProjeto = driver.find_elements_by_xpath("/html/body/main/div[2]/div/div/div[1]/aside/div[1]/a")[0]
btt_novoProjeto.click()

#clicando no campo de busca
wait.until(EC.visibility_of_element_located((By.XPATH, "/html/body/ul/li[16]/a")))
btt_verTodos = driver.find_elements_by_xpath("/html/body/ul/li[16]/a")[0]
btt_verTodos.click()

# preenchendo o campo de busca
pesquisa = driver.find_element_by_id("search")
pesquisa.send_keys("INATEL \_TCC")

# clicando no botão de procurar
btt_procurar = driver.find_element_by_xpath("/html/body/main/div[1]/div[2]/div[1]/button").click()

# clicando no nome do artigo procurado
artigo = driver.find_element_by_id("doc-e31ed62c2b096c560551b734255fe7d8").click()

# abrindo como template
wait.until(EC.visibility_of_element_located((By.XPATH, "/html/body/main/div/div/div[1]/div[2]/div/a[1]")))
template = driver.find_element_by_xpath("/html/body/main/div/div/div[1]/div[2]/div/a[1]").click()

#clicar em figuras
wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="ide-body"]/div[1]/aside/div[2]/ul/file-entity[2]/li/div/div/div'))).click()

#expand 
expand_aba = driver.find_element_by_xpath('//*[@id="ide-body"]/div[1]/aside/div[2]/ul/file-entity[2]/li/div/div/div/span[3]/a').click()

#upload files
upload_files = driver.execute_script("document.querySelector('#ide-body > div.ui-layout-west.ui-layout-pane.ui-layout-pane-west > aside > div.file-tree-inner.ng-scope > ul > file-entity:nth-child(3) > li > div > div > div > span.dropdown.entity-menu-toggle.ng-scope.open > ul > li:nth-child(6) > a').click()")

#selecionando a opção de link externo
driver.execute_script("document.querySelector('body > div.modal.fade.ng-isolate-scope.in > div > div > div.modal-body.modal-new-file.ng-scope > table > tbody > tr > td.modal-new-file--list > ul > li:nth-child(4) > a').click()")

#preenchendo campos
pesquisa_url = driver.find_element_by_name("url")
pesquisa_url.send_keys("https://i.imgur.com/VvOFDIs.jpg")

nome_imagem = driver.find_element_by_name("name")
nome_imagem.clear()
nome_imagem.send_keys("figuraTeste.jpg")

#clicando no botão de pequisa
button_image = driver.execute_script("document.querySelector('body > div.modal.fade.ng-isolate-scope.in > div > div > div.modal-footer.ng-scope > button.btn.btn-primary > span:nth-child(1)').click()")

#expandindo aba 2- introdução
driver.execute_script("document.querySelector('#ide-body > div.ui-layout-west.ui-layout-pane.ui-layout-pane-west > aside > div.file-tree-inner.ng-scope > ul > file-entity:nth-child(4) > li > div > div > div > i.fa.fa-fw.toggle.ng-scope.fa-chevron-right').click()")

#clicando na aba 2 - introdução
driver.find_element_by_xpath('//*[@id="ide-body"]/div[1]/aside/div[2]/ul/file-entity[3]/li/div/ul/file-entity[2]/li/div/div').click()
#driver.find_element_by_xpath('//*[@id="editor"]/div/div[2]/textarea').click()

#desativando auto complete
driver.execute_script("document.querySelector('#chat-wrapper > div.ui-layout-center.ui-layout-pane.ui-layout-pane-center > header > div.toolbar-left > a.btn.btn-full-height > i').click()")
wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="left-menu"]/form/div/div[5]/select'))).click()

escrever_texto = driver.find_element_by_xpath('//*[@id="editor"]/div/div[2]/textarea')

escrever_texto.send_keys("\\begin{figure}\n")
escrever_texto.send_keys("   \\centering\n")
escrever_texto.send_keys("   \\includegraphics[width=\linewidth]{figuras/figuraTeste.jpg}\n")
escrever_texto.send_keys("   \\caption{Rotulo Teste}\n")
escrever_texto.send_keys("   \\label{fig:label_teste}")
escrever_texto.send_keys("\n\\end{figure}\n")

pdf = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="left-menu"]/ul[1]/li[2]/a')))
driver.execute_script("document.querySelector('#recompile > a:nth-child(1)').click()")
wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="recompile"]/a[1]/span[1]')))

pdf = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="left-menu"]/ul[1]/li[2]/a')))
pdf.click()

#voltando ao menu inicial 
driver.execute_script('document.querySelector("#chat-wrapper > div.ui-layout-center.ui-layout-pane.ui-layout-pane-center > header > div.toolbar-left > a.toolbar-header-back-projects > i").click()')

#fazendo filtro para apagar artigos
deletar = driver.find_element_by_xpath("/html/body/main/div[2]/div/div/div[2]/div[2]/div/form/div/input")
deletar.send_keys("Modelo INATEL TCC GRADUACAO")

#Selecionar todos os artigos conforme filtro
driver.execute_script('document.querySelector("body > main > div.project-list-content.ng-isolate-scope > div > div > div.project-list-main.col-md-10.col-xs-9 > div.row.row-spaced > div > div > ul > table > tbody > tr.project-list-table-header-row > th.project-list-table-name-cell > div > input").click()')

#clicando na lixeira para apagar
driver.find_element_by_xpath('/html/body/main/div[2]/div/div/div[2]/div[2]/div/div/div/div[1]/a[3]').click()

#confirmando
#driver.find_element_by_xpath('/html/body/div[2]/div/div/div[3]/button[2]').click()

#driver.quit()
```
